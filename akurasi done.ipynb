# proses tokenisasi bagian preprocessing

```python
#proses tokenisasi untuk preprocessing
#untuk impor nltk
from nltk.tokenize import word_tokenize
tokenisasi_title = [ ]
for title_token in title_en:
    df_token= word_tokenize(title_token)
    title = []
    for c in df_token:
        kecil=c.lower()
#         print(kecil)
        title.append(kecil)
    tokenisasi_title.append(title)

tokenisasi_title
```

# proses filtering(menghilangkan kata tidak penting) bagian proses preprocessing


```python
#proses filtering untuk preprocessing
#fungsi stopword untuk membuang kata-kata yang di rasa tidak penting
# stopword = ['a','to','an','at','for','of','off','and','on','the','in','pada','dan','sebagai','dengan','&','by','=','-',';',
#             '?','.','(',')']

# filtering_title = []
# for title in tokenisasi_title:
#     filt = []
#     for token in title:
#         if token not in stopword:
#              filt.append(token)
#     filtering_title.append(filt)
# filtering_title

import io
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
 
# word_tokenize accepts
# a string as an input, not a file.
stop_words = set(stopwords.words('english'))

filtering_title = []

for tokens in tokenisasi_title:
    words = tokens
    s = []
    for r in words:
        if not r in stop_words:
            s.append(r)
    filtering_title.append(s)
    
filtering_title
```

# proses STEMMING (judul menjadi kata dasar) bagian proses preprocessing

```python
import nltk
from nltk.stem import PorterStemmer, WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

steming_title  = []
word_stemmer = PorterStemmer()
for title in filtering_title:
    stem = []
    for token in title:
#         print(token," : ", word_stemmer.stem(token))
        stem.append(word_stemmer.stem(token))
        lemmatizer.lemmatize(token,pos='v')
        
    steming_title.append(stem)
steming_title
```

```python
#Pembagian data training dan data testing 

XJudul = hasil['stemmed_title'].tolist()
yJudul = hasil['labels'].tolist()

X = dfTFIDF.values.tolist()
y = hasil['labels'].tolist()


xTrain, xTest, yTrain, yTest = train_test_split(X, y, test_size=0.1, random_state=None, shuffle=True, stratify=None)
xTr, xTs, yTr, yTs = train_test_split(XJudul, yJudul, test_size=0.1, random_state=None, shuffle=True, stratify=None)
```

```python
#Algoritma KNN

k = 5

knn = KNeighborsClassifier(n_neighbors=k)
knn.fit(xTrain, yTrain)
label_prediksi = knn.predict(xTest)
judul_prediksi = xTs

dfPrediksi = pd.DataFrame({'Judul':judul_prediksi,'Label':label_prediksi})
dfPrediksi
```

```python

X = np.array(dfTFIDF.values.tolist())
y = np.array(hasil['labels'].tolist())

from sklearn.model_selection import KFold

from sklearn.metrics import confusion_matrix

kf = KFold(n_splits=10)

k = 5

knn = KNeighborsClassifier(n_neighbors=k)

list_akurasi = []
list_precision = []
list_recall = []
list_f1 = []

for train_index, test_index in kf.split(X):    
    
#     print("TRAIN:", train_index, "TEST:", test_index)
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    
    cm = confusion_matrix(y_test, y_pred)
    
    tp= cm[0,0]
    tn= cm[1,1]
    fp= cm[0,1]
    fn= cm[1,0]
    
    akurasi = (tp+tn)/(tp+tn+fp+fn)
    precision = (tp)/(tp+fp)
    recall = ((tp)/(tp+fn))
    f1 = (2*precision*recall)/(precision+recall)
    
    list_akurasi.append(akurasi)
    list_precision.append(precision)
    list_recall.append(recall)
    list_f1.append(f1)
    
dfCM = pd.DataFrame({'Akurasi':list_akurasi, 'Precision':list_precision, 'Recall':list_recall, 'F1':list_f1})
dfCM
    
    
```
